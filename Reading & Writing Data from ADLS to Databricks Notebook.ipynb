# Data Ingestion from ADLS to Databricks Notebooks

###  Reading CSV file

```python
df = spark.read.option("header", True).option("inferschema",True).csv("/mnt/file/employee.csv")
df.display()
```

###  Reading JSON file

```python
df_json = spark.read.option("Multiline", False).option("inferschema",True).json("/mnt/file/people.json")
df_json.display()
```

###  Reading Parquet file

```python
df_parquet = spark.read.option("mergeschema",True).parquet("/mnt/file/userdata1.parquet")
df_parquet.display()
```
